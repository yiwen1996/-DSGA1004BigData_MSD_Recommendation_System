```python
import numpy as np
import pandas as pd
from PIL import Image
import os
import matplotlib.pyplot as plt
```

```python
# Load data: if Windows
image_dir=[]

directory=['min_dist','n_neighbor','spread']

for dir in directory:
    for file in os.listdir(dir):
        image_dir.append(dir+'/'+file)
```

```python
# Parameters for fine tuning
spread = [0.1, 0.5, 0.75, 1, 1.5, 2, 3, 5, 10, 50] # default = 1
n_neighbors = [20, 50, 100, 200, 300, 400, 500, 1000, 1500, 2000] # default = 15
min_dist = [0.00000001, 0.0000001, 0.00001, 0.0001, 0.001, 0.01, 0.05, 0.1, 0.5, 1] # default = 0.1
n_epoch = [] # default = None; 200 for large dataset, 500 for small
```

```python
imagedir_md = image_dir[:10]
imagedir_nn = image_dir[10:20]
imagedir_s = image_dir[20:]
```

```python
max_rows, max_cols = 2, 5
fig, axes = plt.subplots(nrows=max_rows, ncols=max_cols, figsize=(20,8))
for idx, img_dir in enumerate(imagedir_md):
    image = Image.open(img_dir)
    row = idx // max_cols
    col = idx % max_cols
    axes[row, col].axis("off")
    axes[row, col].imshow(image)
    plt.title = 'min_dist = {}'.format(min_dist[idx])
plt.subplots_adjust(wspace=.05, hspace=.05)
plt.savefig('min_dist.png')
plt.show()
```

```python
max_rows, max_cols = 2, 5
fig, axes = plt.subplots(nrows=max_rows, ncols=max_cols, figsize=(20,8))
for idx, img_dir in enumerate(imagedir_nn):
    image = Image.open(img_dir)
    row = idx // max_cols
    col = idx % max_cols
    axes[row, col].axis("off")
    axes[row, col].imshow(image)
    plt.title = 'n_neighbors = {}'.format(n_neighbors[idx])
plt.subplots_adjust(wspace=.05, hspace=.05)
plt.savefig('n_neighbors.png')
plt.show()
```

```python
max_rows, max_cols = 2, 5
fig, axes = plt.subplots(nrows=max_rows, ncols=max_cols, figsize=(20,8))
for idx, img_dir in enumerate(imagedir_s):
    image = Image.open(img_dir)
    row = idx // max_cols
    col = idx % max_cols
    axes[row, col].axis("off")
    axes[row, col].imshow(image)
    plt.title = 'spread = {}'.format(spread[idx])
plt.subplots_adjust(wspace=.05, hspace=.05)
plt.savefig('spread.png')
plt.show()
```
